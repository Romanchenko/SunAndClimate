```python
from collections import defaultdict
import math
# arr - array of values to retrieve entropy
# threshold - which day is supposed to be sunny
# n - we calculate E(n), binary series of length n
# window - window to calculate entropy
def pretty(d):
    res = ''
    for k, v in d.items():
        res += '[ ' + str(k) + ': ' + str(v) + ']; '
    return res

def get_entropy(arr, threshold, n, window):
    m = arr.size
    bin_arr = (arr.values > threshold).astype(int)
    d = {}
    results = np.zeros(m - n + 1)
    p_sum = 0
    for i in range(m - n + 1):
        #print('i=', i, ' d=', pretty(d))
        sub_arr = str(bin_arr[i:i + n])
        #print('sub_arr = ', sub_arr)
        if sub_arr not in d:
            d[sub_arr] = 0
        d[sub_arr] += 1
        if i >= window - n:
            cur_psum, entr = _get_entr(d, window - n + 1)
            p_sum = max(p_sum, cur_psum)
            results[i - window + n] = entr
            sub_arr = str(bin_arr[i - window + n: i - window + n + n ])
            d[sub_arr] -= 1
    print("p_sum = ", p_sum)
    return results


def _get_entr(d, n):
    ans = 0
    denom = n
    p_sum = 0
    for x in d.values():
        p = x / denom
        p_sum += p
        if p > 0:
            ans -= p * math.log(p, 2)
    return p_sum, ans
```

```python
# https://www.weather-atlas.com/en/netherlands/amsterdam-climate#daylight_sunshine
# AVERAGE DAYLIGHT FOR ALL MONTHS:
daylight = {}
daylight[1] = 8.3
daylight[2] = 9.9
daylight[3] = 11.0
daylight[4] = 14
daylight[5] = 15.8
daylight[6] = 16.7
daylight[7] = 16.2
daylight[8] = 14.6
daylight[9] = 12.6
daylight[10] = 10.6
daylight[11] = 8.8
daylight[12] = 7.8
data_rel = data.copy(deep=True)
data_rel['val'] = data_rel.apply(lambda row: row['val']/daylight[row['month']], axis=1)
```

```python
entropys1 = []
threshold = data_rel['val'].mean()
window = 365
for i in range(1, 31):
    print(i, ' in progress')
    entropys1.append(get_entropy(data_rel['val'], threshold, i, window))
```

```python
def explore_simulation(a, c, n):
    x = 1
    xs = []
    plt.figure(figsize=(14, 6))
    for i in range(n):
        y = a * x + np.random.normal(1, 0.8) + c
        xs.append(y)
        x = y
    plt.plot(np.arange(n), xs)
    entropys_sim = []
    window = 365
    xs = np.array(xs)
    threshold = xs.mean()
    for i in range(1, 61):
        print(i, ' in progress')
        entropys_sim.append(get_entropy(xs, threshold, i, window))
    entropys_sim_np = np.zeros((60, len(entropys_sim[0])))
    for i in range(60):
        for j in range(len(entropys_sim[i])): 
            entropys_sim_np[i][j] = entropys_sim[i][j]
    entropys_sim_df = pd.DataFrame(entropys_sim_np)
    return xs, entropys_sim_df
```

```python
def explore_simulation2(a1, a2, c, n):
    x = 1
    xs = []
    m = n // 3
    a = [a1] * m
    plt.figure(figsize=(14, 6))
    for x in np.linspace(a1, a2, m):
        a.append(x)
    a += [a2] * m
    for i in range(m * 3):
        y = a[i] * x + np.random.normal(1, 0.8) + c
        xs.append(y)
        x = y
    plt.plot(np.arange(n), xs)
    entropys_sim = []
    threshold = 0.4
    window = 365
    xs = np.array(xs)
    threshold *= xs.mean()
    for i in range(1, 61):
        print(i, ' in progress')
        entropys_sim.append(get_entropy(xs, threshold, i, window))
    entropys_sim_np = np.zeros((60, len(entropys_sim[0])))
    for i in range(60):
        for j in range(len(entropys_sim[i])):
            entropys_sim_np[i][j] = entropys_sim[i][j]
    entropys_sim_df = pd.DataFrame(entropys_sim_np)
    return xs, entropys_sim_df
```
