```python
from collections import defaultdict
import math
# arr - array of values to retrieve entropy
# threshold - which day is supposed to be sunny
# n - we calculate E(n), binary series of length n
# window - window to calculate entropy
def get_entropy(arr, threshold, n, window):
    m = arr.size
    bin_arr = (arr > threshold).astype(int)
    d = defaultdict(int)
    results = np.zeros(m - n + 1)
    for i in range(m - n + 1):
        sub_arr = str(bin_arr[i:i + n])
        d[sub_arr] += 1
        if i >= window:
            entr = _get_entr(d, window - n + 1)
            results[i] = entr
            sub_arr = str(bin_arr[i - n + 1 : i + 1])
            d[sub_arr] -= 1
    return results


def _get_entr(d, n):
    ans = 0
    denom = n
    for x in d.values():
        p = x / denom
        if p > 0:
            ans -= p * math.log(p, 2)
    return ans
```

```python
# https://www.weather-atlas.com/en/netherlands/amsterdam-climate#daylight_sunshine
# AVERAGE DAYLIGHT FOR ALL MONTHS:
daylight = {}
daylight[1] = 8.3
daylight[2] = 9.9
daylight[3] = 11.0
daylight[4] = 14
daylight[5] = 15.8
daylight[6] = 16.7
daylight[7] = 16.2
daylight[8] = 14.6
daylight[9] = 12.6
daylight[10] = 10.6
daylight[11] = 8.8
daylight[12] = 7.8
data_rel = data.copy(deep=True)
data_rel['val'] = data_rel.apply(lambda row: row['val']/daylight[row['month']], axis=1)
```

```python
entropys = []
threshold = 0.4
window = 365
for i in range(1, 31):
    print(i, ' in progress')
    entropys.append(get_entropy(data_rel['val'], threshold, i, window))
```

```python
df_entropys = np.array(list(map(lambda x : np.array(x), entropys)))
```

```python
def draw_by_pos(pos):
    arr = []
    plt.figure(figsize=(13, 6))
    for x in entropys:
        arr.append(x[pos])
    plt.title('Зависимость прибавки в энтропии от размера последовательности')
    plt.grid(True)
    plt.ylabel('E(t) - E(t-1)')
    plt.xlabel('t')
    plt.plot(arr)
```

```python
draw_by_pos(365)
```
